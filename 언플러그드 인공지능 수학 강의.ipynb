### ANN DNN CNN RNN GAN

1. 퍼셉트론
2. ANN: 인공신경망
3. DNN: ANN의 은닉층을 많이 늘린 것(2개 이상), 오류 역전파 -> 역전파의 원리 학습
4. CNN: 합성곱 신경망, 필터행렬곱 컨볼루션(압축), 풀링, 문장분류 이미지인식 -> 압축 원리 학습
5. RNN: 순환 신경망, 이전 노드 학습결과 다음 노드에 반영, 연속 데이터, 음성이나 자연어처리
6. GAN: 적대적 생성 신경망, 생성기와 판별기가 서로 경쟁하면서 학습.
